```python
import bionetgen
```

```python
res= bionetgen.run("BioModel_ZAP70_SYK_v8_test1.bngl")
```

```python
import matplotlib.pyplot as plt

r= res[0]
names= ['PZAP_total']# r.dtype.names
for name in names:
    if name != "time":
        plt.plot(r['time'], r[name], label = name)
plt.xlabel("time")
plt.ylabel("species (counts)")
_ = plt.legend(frameon = False)
```
